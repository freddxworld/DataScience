```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv("iris.csv")
data = iris_data.head().describe()
```

```python
iris_data.head()
```

```python
#display tuples
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(iris_data)
```

```python
list(iris_data.columns)
```

```python
#display featurs & datatypes
iris_data.info()
```

```python
#display counts
iris_data['Species'].value_counts()
```

```python
#display range
data.loc['range'] = data.loc['max'] - data.loc['min']
print(data)
```

```python
#display scatterplot
sns.scatterplot(data=iris_data, x='sepal_length',
                y='sepal_width', hue='Species')
plt.plot()
```

```python
#display histograms
iris_data.hist()
```

```python
iris_data.boxplot(by="Species", column=["sepal_length", "sepal_width", "petal_length", "petal_width"])
```

```python
sns.pairplot(iris_data, hue="Species")
```

```python
iris_data.boxplot(by="Species", column="sepal_length")
iris_data.boxplot(by="Species", column="petal_length")
iris_data.boxplot(by="Species", column="sepal_width")
iris_data.boxplot(by="Species", column="petal_width")
```
